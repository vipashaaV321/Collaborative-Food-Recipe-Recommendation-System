```python
recipes.rename(columns={'id':'recipe_id'},inplace=True)
recipes.head()
```

```python
ratings_with_name = ratings.merge(recipes, on='recipe_id')
ratings_with_name.head()
```

```python
X = ratings_with_name.groupby('user_id').count()['rating'] > 100
genuine_users = X[X].index
genuine_users
```

```python
# Filtered ratings based on the users...
filtered_ratings = ratings_with_name[ratings_with_name['user_id'].isin(genuine_users)]
```

```python
# Filtering Based on recipes...
y = filtered_ratings.groupby('name').count()['rating']>=50
famous_recipes = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['name'].isin(famous_recipes)]
```

```python
# Dropping Duplicates if any
final_ratings.drop_duplicates()
```

```python
# Creating Pivot Table...
pt = final_ratings.pivot_table(index='name',columns='user_id',values='rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def predict_ratings(user_id):
    user_ratings = pt[user_id]
    user_ratings = user_ratings[user_ratings != 0]
    user_unrated_recipes = pt.index.difference(user_ratings.index)
    predicted_ratings = []
    for recipe_name in user_unrated_recipes:
        similarity_scores = similarity_score[pt.index.get_loc(recipe_name)]
        numerator = (pt[user_id] * similarity_scores).sum()
        denominator = similarity_scores.sum()
        if denominator == 0:
            predicted_rating = 0
        else:
            predicted_rating = numerator / denominator
        predicted_ratings.append((recipe_name, predicted_rating))
    return predicted_ratings
```

```python
def recommend_recipes(user_id):
    predicted_ratings = predict_ratings(user_id)
    top_recipe_names = [recipe_name for recipe_name, predicted_rating in sorted(predicted_ratings, key=lambda x: x[1], reverse=True)[:10]]
    return top_recipe_names
```

```python
recommend_recipes(1533)
```
