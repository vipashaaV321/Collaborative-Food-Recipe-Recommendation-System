```python
# predict the ratings for all uninteracted items for a user
def predict_all(user_id, matrix, item_similarities, k=5):
    # get the ratings of the user
    user_ratings = matrix[user_id, :]

    # get the indices of the uninteracted items
    uninteracted_items = np.where(user_ratings == 0)[0]

    # predict the rating for each uninteracted item
    predictions = np.zeros((len(uninteracted_items),))
    for i, item_id in enumerate(uninteracted_items):
        # find the k most similar items to the target item
        similar_items = np.argsort(-item_similarities[item_id, :])[:k]

        # get the ratings of the k most similar items
        similar_ratings = matrix[user_id, similar_items]

        # compute the weighted average of the ratings
        similarity_scores = item_similarities[item_id, similar_items]
        prediction = np.dot(similarity_scores, similar_ratings) / np.sum(similarity_scores)

        # add the average rating of the item to the prediction
        prediction += np.mean(matrix[:, item_id])

        # store the prediction
        predictions[i] = prediction

    # create a dataframe with the item IDs and predicted ratings
    predicted_ratings = pd.DataFrame({'item_id': uninteracted_items, 'rating': predictions})

    return predicted_ratings
```

```python
import numpy as np

# Define the number of users, items, and latent factors
num_users = 100
num_items = 50
num_factors = 100

# Generate random user-item ratings matrix
R = np.random.randint(0, 6, size=(num_users, num_items))

# Generate a mask matrix indicating which values are missing
missing_mask = np.random.choice([True, False], size=(num_users, num_items), p=[0.3, 0.7])

# Initialize the latent factor matrices U and V with small random values
U = np.random.randn(num_users, num_factors) * 0.01
V = np.random.randn(num_items, num_factors) * 0.01

# Define the regularization parameter
lamda = 0.01

# Define the learning rate and number of iterations
learning_rate = 0.008
num_iterations = 1000

# Define the objective function
def objective_function(R, U, V, lamda, missing_mask):
    error = np.sum(((R - np.dot(U, V.T)) * missing_mask)**2)
    regularization = lamda * (np.sum(U**2) + np.sum(V**2))
    return 0.5 * (error + regularization)

# Perform gradient descent
for iteration in range(num_iterations):
    # Compute the gradients with respect to U and V
    error = (R - np.dot(U, V.T)) * missing_mask
    grad_U = -np.dot(error, V) + lamda * U
    grad_V = -np.dot(error.T, U) + lamda * V
    
    # Update the latent factor matrices U and V
    U -= learning_rate * grad_U
    V -= learning_rate * grad_V
    
    # Compute the objective function and print the progress
    obj = objective_function(R, U, V, lamda, missing_mask)
    print("Iteration {0}/{1}: objective function = {2}".format(iteration+1, num_iterations, obj))
```
